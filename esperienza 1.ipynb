```python
import matplotlib.pyplot as plt
```

```python
volt_1 = [0.234, 0.215, 0.212, 0.210]
amp_1  = [6.9, 130.3, 137.6, 159.0]

volt_2 = [1.013, 0.974, 0.945, 0.483]
amp_2  = [29.3, 2800, 4000, 35500]
```

```python
plt.xlabel(r'$\mu$ A')
plt.ylabel(r'V')
plt.title('batteria 1')

x_error = 0.1
y_error = 0.001

plt.errorbar(amp_1, volt_1, xerr = x_error, yerr = y_error, fmt= '.')
plt.show()
```

```python
plt.xlabel(r'$\mu$ A')
plt.ylabel(r'V')
plt.title('batteria 2')

x_error = [0.1, 100, 100, 100]
y_error = 0.001

plt.errorbar(amp_2, volt_2, xerr = x_error, yerr = y_error, fmt= '.')
plt.show()
```
