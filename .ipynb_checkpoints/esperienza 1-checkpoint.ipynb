```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
volt_1 = np.array([0.234, 0.215, 0.212, 0.210])
amp_1  = np.array([6.9, 130.3, 137.6, 159.0])

volt_2 = np.array([1.013, 0.974, 0.945, 0.483])
amp_2  = np.array([29.3, 2800, 4000, 35500])
```

```python
x_error = 0.1
y_error = np.full((1,4),0.001)
```

```python
w = 1 / y_error**2

# Medie pesate
x_w_mean = np.sum(w * amp_1) / np.sum(w)
y_w_mean = np.sum(w * volt_1) / np.sum(w)

# Calcolo di m (coefficiente angolare pesato)
numerator = np.sum(w * (amp_1 - x_w_mean) * (volt_1 - y_w_mean))
denominator = np.sum(w * (volt_1 - x_w_mean)**2)
m = numerator / denominator

# Calcolo di c (intercetta pesata)
c = y_w_mean - m * x_w_mean

print(f"Coefficiente angolare (m): {m}")
print(f"Intercetta (c): {c}")

# Retta stimata
y_pred = m * amp_1 + c
```

```python
plt.xlabel(r'$\mu$ A')
plt.ylabel(r'V')
plt.title('batteria 1')

plt.errorbar(amp_1, volt_1, xerr = x_error, yerr = y_error, fmt= '.')
plt.plot(amp_1, y_pred, color = 'red')
plt.show()
```

```python
x_error = [0.1, 100, 100, 100]
y_error = np.full((1,4) , 0.001)
```

```python
w = 1 / y_error**2

# Medie pesate
x_w_mean = np.sum(w * amp_2) / np.sum(w)
y_w_mean = np.sum(w * volt_2) / np.sum(w)

# Calcolo di m (coefficiente angolare pesato)
numerator = np.sum(w * (amp_2 - x_w_mean) * (volt_2 - y_w_mean))
denominator = np.sum(w * (volt_2 - x_w_mean)**2)
m = numerator / denominator

# Calcolo di c (intercetta pesata)
c = y_w_mean - m * x_w_mean

print(f"Coefficiente angolare (m): {m}")
print(f"Intercetta (c): {c}")

# Retta stimata
y_pred = m * amp_2 + c
```

```python
plt.xlabel(r'$\mu$ A')
plt.ylabel(r'V')
plt.title('batteria 2')

plt.errorbar(amp_2, volt_2, xerr = x_error, yerr = y_error, fmt= '.')
plt.plot(amp_2, y_pred, color = 'red')
plt.show()
```
