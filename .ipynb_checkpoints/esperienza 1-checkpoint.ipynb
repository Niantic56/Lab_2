```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def MMQ(x,y, y_error):
    w = 1 / y_error**2

    # Medie pesate
    x_w_mean = np.sum(w * x) / np.sum(w)
    y_w_mean = np.sum(w * y) / np.sum(w)

    # Calcolo di m (coefficiente angolare pesato)
    numerator = np.sum(w * (x - x_w_mean) * (y - y_w_mean))
    denominator = np.sum(w * (y - x_w_mean)**2)
    m = numerator / denominator

    # Calcolo di c (intercetta pesata)
    c = y_w_mean - m * x_w_mean

    print(f"Coefficiente angolare (m): {m}")
    print(f"Intercetta (c): {c}")

    # Retta stimata
    y_pred = m * x + c
    
    return y_pred
```

```python
volt_1 = np.array([0.234, 0.215, 0.212, 0.210])
amp_1  = np.array([6.9, 130.3, 137.6, 159.0])

volt_2 = np.array([1.013, 0.974, 0.945, 0.483])
amp_2  = np.array([29.3, 2800, 4000, 35500])
```

```python
x_error = 0.1
y_error = np.full((1,4),0.001)
```

```python
plt.xlabel(r'$\mu$ A')
plt.ylabel(r'V')
plt.title('batteria 1')

plt.errorbar(amp_1, volt_1, xerr = x_error, yerr = y_error, fmt= '.')
plt.show()
```

```python
plt.xlabel(r'$\mu$ A')
plt.ylabel(r'V')
plt.title('batteria 1')

plt.errorbar(amp_1, volt_1, xerr = x_error, yerr = y_error, fmt= '.')
plt.plot(amp_1, MMQ(amp_1, volt_1, y_error), color = 'red')
plt.show()
```
